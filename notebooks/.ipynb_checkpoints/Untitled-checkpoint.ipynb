## 1. Import necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sys
import os
import re

import warnings
warnings.filterwarnings('ignore')
```

## 2. Data Processing

### Reading raw file to dataframe

```python
CURRENT_DIR = os.path.dirname(os.path.abspath('.'))
OUTPUT_PATH_PLANCK = "{}/output/profiling/planck/".format(CURRENT_DIR)
OUTPUT_PATH_RANDOM = "{}/output/profiling/random/".format(CURRENT_DIR)
MAX_HEALPIX_LEVEL = 10
```

```python
df = pd.DataFrame(columns=['level', 'method', 'n_calls', 'total_time', 'per_call', 'cumulative_time', 'per_primitive_call', 'function'])
def process_output_file(filepath, level, method):
    with open(filepath, 'r') as f:
        for num, line in enumerate(f):
            #Skip the first five lines     
            if num >= 5:
                content = re.sub("\s\s+", " ", line).split()
                contents = content[:5]+ [" ".join(content[5:])]
                # Skip empty lines
                if len(contents) == 6:
                    df_length = len(df)
                    df.loc[df_length] = [level, method] + contents
```

```python
for i in range(1, MAX_HEALPIX_LEVEL + 1):
    planck_file = OUTPUT_PATH_PLANCK + "profile_{}.txt".format(i)
    random_file = OUTPUT_PATH_RANDOM + "profile_random_{}.txt".format(i)
    process_output_file(random_file, i, "random")
    process_output_file(planck_file, i, "planck")
```

```python
df.head()
```

### Re-format data

```python
def split_n_calls(n_calls):
    lst = n_calls.split('/')
    if len(lst) == 1:
        return [lst[0], lst[0]]
    else:
        return lst
```

```python
df["total_calls"] = df.apply(lambda row: split_n_calls(row['n_calls'])[0], axis=1)
df["primitive_calls"] = df.apply(lambda row: split_n_calls(row['n_calls'])[1], axis=1)
```

```python
cols = [col for col in list(df.columns) if col not in ['function', 'n_calls', 'method']]
for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

## 3. Profiling Analysis

### 3.1. Function call

```python
df_group = df.groupby(['method', 'level'])
```

```python
df_call = df_group['total_calls', 'primitive_calls'].sum().reset_index()
```

```python
df_call
```

```python
x = list(df_call[df_call['method'] == 'planck']['level'])
y_planck = [np.log(x) for x in list(df_call[df_call['method'] == 'planck']['total_calls'])]
y_random = [np.log(x) for x in list(df_call[df_call['method'] == 'random']['total_calls'])]

fig, ax = plt.subplots()

plt.plot(x, y_planck, label = "planck")
plt.plot(x, y_random, label = "random")
#plt.title("Total function calls")

ax.set_ylabel('Total calls (log)')
ax.set_xlabel('Level')
plt.legend()
plt.show()
```

```python
df_call['non_primitive_calls'] = df_call['total_calls'] - df_call['primitive_calls'] 
y_planck = list(df_call[df_call['method'] == 'planck']['non_primitive_calls'])
y_random = list(df_call[df_call['method'] == 'random']['non_primitive_calls'])

fig, ax = plt.subplots()

plt.plot(x, y_planck, label = "planck")
plt.plot(x, y_random, label = "random")

ax.set_ylabel('Total calls')
ax.set_xlabel('Level')

plt.title("Total non-primitive function calls")
plt.legend()
plt.show()
```

### 3.2. Running time 

#### Random data:

```python
df1 = df[(df['method'] == 'random') & (df['function'] == 'evaluation.py:13(generate_random_sources)')]
df1
```

```python
df2 = df[(df['method'] == 'random') & (df['function'] == 'conjugate_solution.py:81(findSolution)')]
df2
```

```python
y_generate = [np.log(x) for x in list(df1['cumulative_time'])]
y_calculate = [np.log(x) for x in list(df2['cumulative_time'])]

fig, ax = plt.subplots()

plt.plot(x, y_generate, label = "Generating data")
plt.plot(x, y_calculate, label = "Finding solution")
plt.title("Time spent on random data evaluation (log)")

ax.set_ylabel('Log time')
ax.set_xlabel('Level')
plt.legend()
plt.show()
```

```python
labels = x
y_generate = list(df1['cumulative_time'])
y_calculate = list(df2['cumulative_time'])
y_total = [y_generate[i] + y_calculate[i] for i in range(len(y_generate))]
width = 0.35

fig, ax = plt.subplots()

ax.bar(labels, y_generate, width, label='Generating data')
ax.bar(labels, y_calculate, width, bottom=y_generate, label='Finding Solution')
ax.plot(labels, y_total, label='Total time')
ax.plot(labels, )

ax.set_ylabel('Time (s)')
ax.set_title('Time spent on random data evaluation')
ax.legend()

plt.show()
```
